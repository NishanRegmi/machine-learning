```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(41)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=41)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=41)
```

```python
dt.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 6))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = dt.predict(np.c_[xx1.ravel(), xx2.ravel()]).reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.title("Decision tree")
plt.show()
```

In above decision tree graph, we can clearly see that the model is overfit and perform very well on training data but poor in unseen (new) data.So to tackle this issue, we apply for random forest classifier.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500, random_state=41)
rf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 5))
x_ragne = np.linspace(X.min(), X.max(), 100)
xx, yy = np.meshgrid(x_ragne, x_ragne)
y_pred = rf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, y_pred, alpha=0.2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.title("Random Forest")
plt.show()
```

```python
n_train = 150
n_test = 1000
noise = 0.1

#generate data
def f(x):
    return np.exp(-x**2) + 1.5 *np.exp(-(x-2) **2)

def generate(n_samples, noise):
    X = np.random.rand(n_samples)*10 - 5
    X = np.sort(X).ravel()
    y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2)\
        + np.random.normal(0.0, noise, n_samples)
    X = X.reshape((n_samples, 1))
    
    return X, y
```

```python
X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(X_test, f(X_test), 'r')
plt.scatter(X_train, y_train, c='b')
plt.xlim(-5, 5)
plt.show()
```

```python
#one decision tree regressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(X_train, y_train)
y_pred0 = dtree.predict(X_test)

plt.figure(figsize=(10, 5))
plt.plot(X_test, f(X_test), 'r', label='Original')
plt.scatter(X_train, y_train, c='b')
plt.plot(X_test, y_pred0, 'g', label='Decision Tree')
plt.xlim(-5, 5)
plt.title('Decision Tree score')
plt.legend()
plt.show()
```

```python
print("accuracy: ", r2_score(y_test, y_pred0))
```

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
y_pred1 = rfr.predict(X_test)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(X_test, f(X_test), 'r', label = 'original')
plt.scatter(X_train, y_train, c='b')
plt.plot(X_test, y_pred1, 'g', label = 'random forest')
plt.xlim(-5, 5)
plt.legend()
plt.title("Random Forest")
plt.show()
```

```python
print("Accuracy: ", r2_score(y_test, y_pred1))
```
