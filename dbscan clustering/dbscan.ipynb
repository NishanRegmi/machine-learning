```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
```

```python
X, y = make_circles(n_samples=200, noise=0.03, factor=0.5, random_state=41)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
#applying dbscan to this dataset
dbscan = DBSCAN(min_samples=5, eps=0.1)
cluster = dbscan.fit_predict(X)
```

```python
#plot
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='rainbow')
plt.title("DBSCAN")
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```

**Advantages**

i. robust to outliers

ii. no need to specify cluster number

iii. can find arbitrary shaped cluster

**Disadvantages**

i. sensetive to hyperparameters

ii. difficulty with varying density cluster

iii. does not predict
