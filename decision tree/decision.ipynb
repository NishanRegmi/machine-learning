```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
def tree_analyzer(depth):
    df = pd.read_csv('Social_Network_Ads.csv')
    X = df.iloc[:, 2:4].values
    y = df.iloc[:, -1].values

    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(X, y)
    
    a = np.arange(start=X[:, 0].min()-1, stop=X[:, 0].max()+1, step=0.1)
    b = np.arange(start=X[:, 1].min()-1, stop=X[:, 1].max()+1, step=100)

    xx, yy = np.meshgrid(a, b)
    inpup_array = np.array([xx.ravel(), yy.ravel()]).T

    labels = clf.predict(inpup_array)

    plt.contourf(xx, yy, labels.reshape(xx.shape), alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    return plt.show()
```

```python
tree_analyzer(depth= 4)
```
