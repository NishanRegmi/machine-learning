```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ushape.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, 0:2].values
y = df.iloc[:, -1].values
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = 'winter')
plt.show()
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X, y)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X, y.astype('int'), clf, legend=2)
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(clf, X, y, scoring='accuracy', cv=10))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=3, include_bias=False)
X_trf1 = poly.fit_transform(X)
```

```python
type(X_trf1)
```

```python
clf1 = LogisticRegression()
clf1.fit(X_trf1, y)
```

```python
np.mean(cross_val_score(clf1, X_trf1, y, scoring='accuracy', cv=10))
```

```python
def decision_boundary(X, y, degree=1):
    poly = PolynomialFeatures(degree=degree)
    X_trf = poly.fit_transform(X)

    lr = LogisticRegression()
    lr.fit(X_trf, y)

    accuracy = np.mean(cross_val_score(lr, X_trf, y, scoring='accuracy', cv=10))

    a=np.arange(start=X[:,0].min()-1, stop=X[:,0].max()+1, step=0.01)
    b=np.arange(start=X[:,1].min()-1, stop=X[:,1].max()+1, step=0.01)


    XX,YY=np.meshgrid(a,b)
    
    input_array=np.array([XX.ravel(),YY.ravel()]).T

    labels=lr.predict(poly.transform(input_array))
    
    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha=0.5)
    plt.scatter(X[:,0],X[:,1], c=y)
    plt.title('Degree = {}, accuracy is {}'.format(degree,np.round(accuracy,4)))
```

```python
decision_boundary(X, y, degree=3)
```
