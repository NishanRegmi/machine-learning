```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts(ascending=False, dropna=False).plot(kind='bar')
plt.show()
```

```python
df['GarageQual'].mode()
```

```python
plt.figure(figsize=(8, 5))
plt.subplot(111)

sns.kdeplot(df[df['GarageQual'] == 'TA']['SalePrice'], color='blue', label = 'saleprice with TA')
sns.kdeplot(df[df['GarageQual'].isnull()]['SalePrice'], color='red', label = 'saleprice with NA')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde', ax=ax)

df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Houses with TA', 'Houses with NA']
ax.legend(lines, labels, loc='best')

plt.title('GarageQual')
```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA', inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


temp.plot(kind='kde', ax=ax)

# distribution of the variable after imputation
df[df['GarageQual'] == 'TA']['SalePrice'].plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()
labels = ['Original variable', 'Imputed variable']
ax.legend(lines, labels, loc='best')

# add title
plt.title('GarageQual')
```
