```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_diabetes
```

```python
X, y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr', lr), ('dt', dt), ('svr', svr)]
```

```python
for estimator in estimators:
    x = cross_val_score(estimator[1], X, y, cv=10, scoring='r2')
    print(estimator[0], np.round(np.mean(x), 2))
```

```python
from sklearn.ensemble import VotingRegressor
vr = VotingRegressor(estimators=estimators)

x = cross_val_score(vr, X, y, cv=10, scoring='r2')
print(np.round(np.mean(x), 2))
```

```python
for i in range(1, 3):
    for j in range(1, 3):
        for k in range(1, 3):
            vr = VotingRegressor(estimators=estimators, weights=[i, j, k])
            x = cross_val_score(vr, X, y, cv=10, scoring='r2')
            print("for i={}, j={}, k={}".format(i, j, k), np.round(np.mean(x), 2))
```

```python
#using the same algorithm
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=2)
dt3 = DecisionTreeRegressor(max_depth=3)
dt4 = DecisionTreeRegressor(max_depth=4)
dt5 = DecisionTreeRegressor(max_depth=5)

estimate = [('dt1', dt1), ('dt2', dt2), ('dt3', dt3), ('dt4', dt4), ('dt5', dt5)]
```

```python
for i in estimate:
    x = cross_val_score(i[1], X, y, cv=10, scoring='r2')
    print(i[0], np.round(np.mean(x), 2))
```

```python
#voting regressor
vr = VotingRegressor(estimators=estimate)
score = cross_val_score(vr, X, y, cv=10, scoring='r2')
print('Score', np.round(np.mean(score), 2))
```
